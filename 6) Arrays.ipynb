# Arrays: an array is an object that stores a collection of elements of the same data type.

While similar to lists, arrays in Python are more memory-efficient and perform faster when dealing with large numbers of elements of the same type. Unlike lists, which can hold elements of different data types, arrays are constrained to a single type.

# creating an array:
arrayname = array(type code, [elements]) 


```python
# importing array module
# method 1
import array
a=array.array("i",[23,25,1,4,45,78])
for i in a:
    print(i)
print(a)
```

# Array Indexing:
#### Array indexing is the process of accessing or retrieving a specific element from an array by its position number. In Python, array elements are assigned a position number starting from 0. 
Range: o to n-1

```python
import array
# Create an array of integers with type code 'i'
my_array = array.array('i', [10, 20, 30, 40, 50])

# Access elements using positive indices
print("Element at index 0:", my_array[0])
print("Element at index 2:", my_array[2])
print("Element at index 4:", my_array[4])

# Access elements using negative indices
print("\nElement at index -1 (last element):", my_array[-1])
print("Element at index -3:", my_array[-3])
```

# Array Slicing:
#### Array slicing is the process of extracting a contiguous piece of an array, allowing you to work with a subset of its elements. It's a powerful feature for retrieving a range of elements at once, unlike indexing, which retrieves a single element.
syntax: array_name[start:stop:stepsize]

```python
import array

# Create an array of integers
my_array = array.array('i', [10, 20, 30, 40, 50, 60, 70])
print("Original array:", my_array)

# Slice to get elements from index 1 to 3 (exclusive of index 4)
slice1 = my_array[1:4]
print("Slice from index 1 to 3:", slice1)
# Output: array('i', [20, 30, 40])

# Slice to get every other element from the beginning
slice2 = my_array[::2]
print("Slice with step size 2:", slice2)
# Output: array('i', [10, 30, 50, 70])

# Slice the array in reverse order
slice3 = my_array[::-1]
print("Reversed slice:", slice3)
# Output: array('i', [70, 60, 50, 40, 30, 20, 10])

# You can also slice and update an array
my_array[1:3] = array.array('i', [95, 96])
print("Array after update:", my_array)
# Output: array('i', [10, 95, 96, 40, 50, 60, 70])
```
