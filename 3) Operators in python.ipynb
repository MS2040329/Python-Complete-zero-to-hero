# Operators in Python:
An operator in Python is a symbol that performs an operation on one or more values, which are called operands.
# Types of operators in Python:

## Arithmetic Operators: 
Used for basic mathematical calculations like addition (+), subtraction (-), multiplication (*), and division (/). They also include modulus (%), exponentiation (**), and integer division (//).

## Assignment Operators: 
Used to assign values to variables. The equals sign (=) is the primary assignment operator, and compound assignment operators like += and -= combine an arithmetic operation with assignment.

## Relational Operators: 
Used to compare two values, returning either True or False. Examples include greater than (>), less than (<), and equal to(==).

## Logical Operators:
Used to combine conditional statements and create compound conditions. These include and, or, and not.

## Bitwise Operators:
Perform operations on the individual bits of numbers. They include bitwise AND (&), OR (|), and XOR (^), as well as left shift (<<) and right shift (>>).

## Membership Operators:
Used to test for the presence of a value within a sequence (like a string, list, or tuple). The two operators are 
in and not in.

## Identity Operators:
Used to compare the memory locations of two objects to see if they are the same. The operators are 
is and is not.

## Boolean Operator:
In Python, boolean operators are a set of operators that act on bool type literals, which are either True or False.
and, or, and not

## Unary minus operator:
The unary minus operator is denoted by the symbol minus ( - ). When this operator is 
used before a variable, its value is negated. 

```python
# Arithematic operators

# Variables
a = 15
b = 4

# Addition
print("Addition:", a + b)  

# Subtraction
print("Subtraction:", a - b) 

# Multiplication
print("Multiplication:", a * b)  

# Division
print("Division:", a / b) 

# Floor Division
print("Floor Division:", a // b)  

# Modulus
print("Modulus:", a % b) 
# Exponentiation
print("Exponentiation:", a ** b)
```

```python
#Assignment operator:
a = 10
b = a
print(b)
b += a
print(b)
b -= a
print(b)
b *= a
print(b)
b <<= a
print(b)
```

```python
# Relational operators

a = 13
b = 33

print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

```python
# logical operators

a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

```python
# Bitwise operator
a = 10
b = 11

# Bitwise and(&) ( ampersand)
# a= 10 = 0000 1010.  
# b= 11 = 0000 1011.
# a&b =   0000 1010  This is nothing but 10 (in decimal). 
print(a & b)  

# Bitwise or(|) ( pipe symbol)
# a = 10 = 0000 1010.  
# b = 11 = 0000 1011.
#    a|b = 0000 1011  This is nothing but 11 (in decimal). 
print(a | b) 

#Bitwise not(~) (tilde)
#By changing 0’s to 1’s and vice versa, we get 1111 0101. This is nothing but -11(in decimal). So, ~x = -11. 
print(~a)

# Bitwise xor(^) (cap, carat, or circumflex symbol.) (odd no of 1's)
# a = 10 = 0000 1010.  
# b = 11 = 0000 1011.
#    a^b = 0000 0001
print(a ^ b) 

# Bitwise Left Shift Operator (<<) (double less than)
# Shifting the value of x to the left 2 positions will cause the leftmost 2 bits to be lost.
# The value of x is 10 = 0000 1010. Now, x<<2 will be 0010 1000 = 40 (in decimal).
# To perform a left shift of a number n by x positions (i.e., n << x), you simply divide the number by 2^x.
print(a >> 4) 

# Bitwise Right Shift Operator (>>) (double greater than)
# Sign bit 0 represents a positive number and 1 represents a negative number. So, after performing >> operation on a positive number, 
# we get a positive value in the result also. If right shifting is done on a negative number, again, we get a negative value only. 
# Shifting the value of x to the right 2 positions will cause the rightmost 2 bits to be lost. 
# x value is 10 = 0000 1010. Now x>>2 will be: 0000 0010 = 2 (in decimal) 
# To perform a right shift of a number n by x positions (i.e., n >> x), you perform integer multiply of the number by 2^x.
print(a << 2) 
```

```python
# Membership operators(in and not in)
# in operator: returns True if an element is found in the specified sequence. If the element is not found in the sequence, then it returns False.
# not in operator: does the reverse of the in operator.

for i in range(0,10):
    print(i)
```

```python
# Identity operator:
# These operators compare the memory locations of two objects. Hence, it is possible to 
# know whether the two objects are same or not. The memory location of an object can be 
# seen using the id() function. This function returns an integer number, called the identity 
# number that internally represents the memory location of the object. 
a=10
b=10
print(id(a))
print(id(b))
```

```python
# Identity operators:
# The is Operator  
# The ‘is’ operator is useful to compare whether two objects are the same or not. It will 
# internally compare the identity number of the objects. If the identity numbers of the 
# objects are the same, it will return True; otherwise, it returns False. 
a = 25 
b = 25 
if(a is b): 
    print("a and b have same identity") 
else: 
    print("a and b do not have same identity") 
# The is not Operator 
# This is not operator returns True, if the identity numbers of two objects being compared 
# are not same. If they are same, then it will return False.
one = [1,2,3,4] 
two = [1,2,3,4] 
if(one is not two): 
    print("one and two are same") 
else: 
    print("one and two are not same") 
```

# Associativity of operators:
Associativity is a rule that determines the order of evaluation for operators that have the same precedence level in an expression. It specifies whether an expression is evaluated from left to right or from right to left.

# Mathematical Functions in Python:
Mathematical functions in Python are predefined functions that perform various mathematical calculations. These functions are primarily found in the math module, a standard part of Python's library. To use them, you must first import the math module into your program.





```python
import math as m
#using ceil(x):Raises x value to the next higher integer. If x is an integer, then the same value is returned.
x=4.56
print(m.ceil(x))

#floor(x): Decreases x value to the previous integer value. If x integer, then same value is returned.
print(m.floor(x))

#degrees(x): convert angle value x from radian to degree
print(m.degrees(x))

#radians(x): convert angle value x from degree to radian
x=180
print(m.radians(x))

#sin(x),cos(),tan():returns sin cos and tan values of x in radians
print(m.sin(x))
print(m.cos(x))
print(m.tan(x))

#exp(x):Returns exponentiation of x. This is the same as e **x.
x=10
print(m.exp(x))

#fabs(x): gives absolute value of x
print(m.fabs(x))

#factorial(x):Returns factorial value of x. Raises ‘ValueError’ if x is not an integer or is negative.
print(m.factorial(x))

#fmod(x,y):Returns remainder of division of x and y. works for float values also
print(m.fmod(11,10))

#modf():Returns float and integral parts of x. 
print(m.modf(10.5))

#fsum(): Returns an accurate sum of floating point values.  
print(m.fsum([2.34,4.25,15.69,23.56]))

# log10(x):Returns base-10 logarithm of x
print(m.log10(x))

# log(x,base):returns natural log of x with specified base
print(m.log(x,2))

#sqrt(x): find square root of x
x=100
print(m.sqrt(x))

#pow(x,y): 
x=5
y=3
print(m.pow(x,y))

#gcd(x,y): finds greatest common divisor of x and y
print(m.gcd(50,100))

#trunc(x): The real value of x is truncated to an integer value and returned.
print(m.trunc(23.48597615489))

# isinf(): Returns True if x is a positive or negative infinity, and False otherwise. 
print(m.isinf(m.inf))

#isnan(): Returns True if x is a NaN (not a number), and False otherwise.
print(m.isnan(m.nan))

print(m.e)
print(m.pi)
```
