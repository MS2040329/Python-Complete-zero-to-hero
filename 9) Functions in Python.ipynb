```python
# a function to test whether a number is prime or not 
def prime(n): 
    """ to check if n is prime or not """ 
    x=1   # this will be 0 if not prime 
    for i in range(2, n):   # divide n from 2 to n-1 
        if n%i == 0:   # if divisible by any number 
            x=0   # take x as 0 
            break 
        else: 
            x=1   # else take x as 1 
    return x 
 
# test if a number is prime or not 
num = int(input('Enter a number: ')) 
 
# check if num is prime or not 
result = prime(num) 
if result == 1: 
    print(num, ' is prime') 
else: 
    print(num, ' is not prime')
```

# Types of Actual arguments:
## Positional arguments: 
### These are the arguments passed to a function in correct positional order.
## Keyword arguments: 
### These arguments are passed to a function by specifying the parameter name followed by an equals sign and the value (parameter_name=value). The main advantage is that their order does not matter.
## Default arguments: 
### A default argument is used when a function is called without providing a value for a specific parameter. The function definition provides a default value for that parameter.
## Variable lenght arguments:
### These are used when you don't know in advance how many arguments will be passed to a function.
### __*args__: This collects any number of positional arguments into a tuple.

### __**kwargs__: This collects any number of keyword arguments into a dictionary..  

```python
# positional arguments demo 
def attach(s1, s2): 
    """ to join s1 and s2 and display total string """ 
    s3 = s1+s2 
    print('Total string: '+s3) 
# call attach() and pass 2 strings 
attach('New', 'York')  # positional arguments

def greet(name, message):
  print(f"{message}, {name}!")

# 'Manoj' is passed to 'name', 'Hello' is passed to 'message'
greet('Manoj', 'Hello')
```

# Recursive functions: A function that calls itself is known as â€˜recursive functionâ€™.

```python
# program to calculate the factorial of a number using function
def fact(n):
    if n==0:
        result=1
    else:
        result= n*fact(n-1)
    return result
for i in range(1,11):
    print(fact(i))
```

```python
def hanoi(n, source, target, auxiliary):
    # Base case: only one disk
    if n == 1:
        print(f"Move disk 1 from pole {source} to pole {target}")
    else:
        # Move n-1 disks from source to auxiliary using target as temporary
        hanoi(n - 1, source, auxiliary, target)
        # Move the nth disk from source to target
        print(f"Move disk {n} from pole {source} to pole {target}")
        # Move n-1 disks from auxiliary to target using source as temporary
        hanoi(n - 1, auxiliary, target, source)

# Input from user
n = int(input("Enter number of disks: "))
hanoi(n, 'A', 'C', 'B')  # A = source, C = target, B = auxiliary
```

# Decorators in Python:
## A decorator is a function that takes another function as a parameter, adds some functionality to it, and then returns a function. Decorators are a powerful feature that allows you to modify or enhance the behavior of a function without permanently changing its source code.
## To apply the decorator to any function, we can use the â€˜@â€™ symbol 


```python
# 1. Defining the decorator
def my_decorator(func):
  def wrapper():
    print("--- Starting Decoration ---")
    func() # Calling the original function
    print("--- Ending Decoration ---")
  return wrapper

# 2. Applying the decorator using the @ syntax
@my_decorator
def say_hello():
  print("Hello!")

# 3. Calling the decorated function
say_hello()
```

```python
# decorator to increase the value of a function by 2
def decor(fun):
    def wrapper():
        value=fun()
        return value+2
    return wrapper
@decor
def num():
    return 10
num()
```

```python
# program to create 2 decorators:
def decor(func):
    def inner():
        value=func()
        return value+2
    return inner
def decor1(func):
    def inner():
        value=func()
        return value*2
    return inner
@decor
@decor1
def num():
    return 10
print(num())

# callnum() function and apply decor1 and then decor  
result_fun = decor(decor1(num)) 
print(result_fun()) 
```

# Generators in Python:
## Generators are special functions that return a sequence of values one at a time. Instead of building an entire list in memory, a generator "yields" one value, pauses its execution, and then resumes from where it left off the next time a value is requested.

##  This makes them very memory-efficient, especially when working with large datasets.

```python
def num(x,y):
    while x<=y:
        yield x
        x+=1
generator=num(5,10)
for i in generator:
    print(i,end=' ')
generator=num(11,20)
lst=list(generator)
print(lst)
```

```python
def num(x,y):
    while x<=y:
        yield x
        x+=1
generator=num(5,10)
lst=list(generator)
print(lst)
```

```python
def my_char():
    yield 'A'
    yield 'B'
    yield 'C'
g=my_char()
print(next(g))
print(next(g))
print(next(g))
print(next(g)) # returns error When all values have been yielded, calling next() again will raise a StopIteration exception.
```

# __Structured/Functional programming__:
## It is a problem-solving technique used in various areas, the approach is to break the complex problems into smaller problems, more manageable tasks called sub-tasks. Each of these subtasks is then implemented as several functions. 

## This makes main function easy to read and debug of the functions.

```python
# example of structural programming
# calculating the salary of an employee
basic=float(input('enter basic salary of employee'))
def da(basic):
    da=basic*80/100  # da is 80% of basic salary
    return da
def hra(basic):
    hra = basic*15/100 # hra is 15% of basic salary
    return hra
def pf(basic):
    pf=basic*12/100
    return pf
def itax(gross):
    tax=gross*0.1
    return tax
gross=basic+da(basic)+hra(basic)
print('{:10.4f}'.format(gross))
net=gross-pf(basic)-itax(gross)
print('{:10.4f}'.format(net))
```

# Special variable \__name__ :
##  This variable stores information regarding whether the program is executed as an individual program or as a module.
## When the program is executed directly, the Python interpreter stores the value â€˜__main__â€™ into this variable.
## When the program is imported as a module into another program, then the Python interpreter stores the module name into this variable. Thus, by observing the value of the variable â€˜__name__â€™, we can understand how the program is executed.ted..

```python
# Save this as one.py
def display():
  print('Hello Python!')

# This block checks how the script is being run
if __name__ == '__main__':
  display() # Call display function
  print('This code is run as a program')
else:
  print('This code is run as a module')
```
