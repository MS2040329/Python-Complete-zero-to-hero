# Regular Expression:
## A regular expression is a string that contains special symbols and characters to find and extract the information needed by us from the given data. 
## A regular expression helps us to search for information, match, find and split information as per our requirements. 
## A Regular expression is also called simply regex. 
## Python provides re module that stands for regular expressions. This module contains methods like compile(), search(), match(), findall(), split(), etc.  

```python
import re 
prog = re.compile(r'm\w\w') # compile the regular expression using compile() method
str = 'cat mat bat rat'  
result = prog.search(str) # search the regular expression on the string
print(result.group()) # display the result using group() method
```

```python
str1 = 'Operating system met format' 
result = prog.search(str1) #prog holds the compiled regex. You donâ€™t need to call re.compile again for each string. This improves the execution speed.
print(result.group()) 
```

```python
str='met mop met '
result=re.search("m\w\w",str)
if result:
    print(result.group())
```

```python
result=re.findall("m\w\w",str) # findall() method find all the instances of the regular expression in the string.
print(result) #It returns the result in the form of a list.
for _ in result:
    print(_)
```

```python
result=re.match("m\w\w",str) #match() that returns the resultant string only if it is found in the beginning of the string. 
print(result.group())
```

```python
# The match() method will give None if the string is not in the beginning.
str='not mat welcome'
result=re.match("m\w\w",str)
print(result)
```

```python
str='not mat welcome'     # re.split() splits the string into pieces based on the regular expression
result=re.split(r"m\w\w",str)  # Returns the result in the form of the list
print(result)
```

```python
# find a string and replace it with a new string
str= 'Python is bad for programming'
result=re.sub(r"bad","good",str)
print(result)
```

# __Sequence characters in Regular Expression__: Used to match for a single character in the string



```python
str='an apple a day keeps the doctor away'
result=re.findall(r"a[\w]*",str)
for _ in result:
    print(_)
```

```python
import re
str='an apple a day keeps the doctor away'
result=re.findall(r"\ba[\w]*\b",str)
for words in result:
    print(words)
```

```python
str='the college will reopen on 5th of january'
result=re.findall(r"\d[\w]*",str)
for i in result:
    print(i)
```

```python
#A character in curly braces, e.g. {m} represents repetition for m times. 
str = 'one two three four five six seven 8 9 10'  
result = re.findall(r'\b\w{5}\b', str) 
print(result) 
```

```python
result = re.search(r'\b\w{5}\b', str) 
print(result.group()) 
```

```python
str = 'one two three four five six seven 8 9 10'  
result = re.findall(r'\b\w{4,}\b', str) 
print(result) 
```

```python
# {3,5} indicates 3 to 5 characters in the string  
result = re.findall(r'\b\w{3,5}\b', str) 
print(result) 
```

```python
result = re.findall(r'\b\d\b', str) 
print(result) 
```

```python
str = 'one two three one two three'  
result = re.findall(r't[\w]*\Z', str) # \z represents searching should be done at the end of the string
print(result) 
```

```python
str = 'the two three one two three'  
result = re.findall(r'\At[\w]*', str) 
print(result) 
```

# Quantifiers in regular expression:
## some characters represent more than one character to be matched in the string. Such characters are called â€˜quantifiersâ€™.
.

```python
str='Manoj: 8059460773'
result=re.search(r"\d+",str)
result1=re.search(r"\D+",str)
print(result.group())
print(result1.group())
```

```python
str='manoj manohar mohinder anil manish mehta'
result=re.findall(r'm[ao]\w+',str) #[ao] represents a or o
for i in result:
    print(i)
```

```python
# to retrieve the date of birth from a string
str = 'Vijay 20 1-5-2001, Rohit 21 22-10-1990, Sita 22 15-09-2000'
result=re.findall(r'\d{1,2}-\d{1,2}-\d{4}',str)
print(result)
```

# __Special characters in regular expression__:


```python
str = "Hello World" 
result = re.search(r"^He", str) 
if result: 
    print ("String starts with 'He'") 
else: 
    print("String does not start with 'He'") 
```

```python
result=re.search(r"World$",str)
if result:
    print("string ends with World")
else:
    print("string does not ends with 'World'")
```

# Note: `Regular expressions conduct a case sensitive searching for the strings.`

```python
# To avoid cases of characters in searching we use IGNORECASE of the regular expression module
str='Hello World'
result=re.search(r'world$',str,re.IGNORECASE)
print(result.group())
```

##  A hyphen (-) represents a range of characters. For example, [A-Z] represents range of characters from A to Z

```python
# retrieve names and marks from the string
str="Manoj:89 , Chander:56 , Mohit:89 , Pankaj:45"
result=re.findall(r"[A-Z][a-z]*",str)
result1=re.findall(r"\d{2}",str)
for _ in result:
    print(_)
for _ in result1:
    print(_)
```

```python
# Regular Expression on files:
f=open("Practice.txt", "r")
for line in f:
    result=re.findall('[\w+]*@gmail.com',line)
    for i in result:
        print(i)
f.close()
```

```python
f1=open("Practice.txt", "r")
f2=open("New.txt", "w")
for line in f1:
    result=re.findall('[\w+]*@gmail.com',line)
    for i in result:
        f2.write(i+'\n')
f1.close()
f2.close()
```
