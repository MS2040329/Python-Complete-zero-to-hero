# What is OOPS?
## Object-Oriented Programming (OOP) is a programming approach that uses classes and objects to structure a program. Instead of focusing on procedures or functions, OOP organizes a complex task by breaking it down into self-contained modules called classes. 

# What is procedural-oriented programming?
## While developing software, the main task is divided into several subtasks, and each subtask is represented as a procedure or function. This approach is called procedural-oriented programming.
 

# Problems of Procedural-oriented programming:
## __Lack of Reusability__: In a procedural approach, code is often not designed to be reusable. When a new task is required, it frequently means developing new functions from scratch, which wastes time and effort. 

## __Difficult Maintenance and Debugging__: Since the program is a collection of interdependent functions, an error in one function can have cascading effects. This means a programmer might have to examine all related functions to find and fix a single bug, making updates and debugging difficult. 

## __Poor Scalability__: As the size of the program grows (e.g., beyond 10,000 lines of code), programmers can start to lose control, finding it hard to understand the code's exact behavior. This makes the procedural approach unsuitable for very large and complex systems. 

## __Not Modeled on the Real World__: The approach is based on functions and procedures, which are abstract concepts. It isn't modeled on how humans perceive the real world (as a collection of objects), making it a less natural or intuitive way to program.

# What is class?
## A class is a model or plan used to create objects. It defines a set of common attributes (variables) and actions (methods) that all objects of that class will possess. A class is a logical concept and doesn't physically exist in memory in the same way an object does; it's just the blueprint.

# What is an object?
## An object is an instance of a class. It's a real entity that is created based on the class blueprint and takes up space in memory. An object holds the actual data for the attributes defined in its class. You can create many different objects from the same class, each being a unique instance.

# Encapsulation in OOPS:
## Encapsulation is a core OOP principle where the data (variables) and the code that operates on that data (methods) are bundled together into a single unit. In Python, the class itself is the primary example of encapsulation.
## The variables and methods of a class are called members of the class.
## While a class bundles data and methods, encapsulation also involves the concept of data hiding. Python's approach is slightly different from languages like Java or C++.

## __Public by Default__: All attributes and methods in a Python class are public by default, meaning they can be accessed from outside the class.

## __Simulating Private Members__: To indicate that an attribute should be treated as private (not for external use), you can prefix its name with double underscores (__). This invokes a mechanism called __name mangling__, which makes the attribute harder to access accidentally from outside the class.

## Name mangling is a process in Python where the interpreter changes the name of a class attribute that is prefixed with two underscores (__). This is done to avoid accidental name collisions with attributes in subclasses and to signal that the attribute is intended for internal use only. It's Python's way of simulating private variables.

```python
class Student: 
# to declare and initialize the variables 
    def __init__(self): 
        self.id = 10 
        self.name = 'Raju' 
# display students' details 
    def display(self): 
        print(self.id) 
        print(self.name) 
s1=Student()
s1.display()
```

```python
# A class with instance variables and an instance method
class Student:
  # The constructor to initialize instance variables
  def __init__(self):
    self.name = 'Manoj Sharma'
    self.age = 20
    self.marks = 900

  # An instance method to display the data
  def talk(self):
    print('Hi, I am', self.name)
    print('My age is', self.age)
    print('My marks are', self.marks)

# Create an instance (object) of the Student class
s1 = Student()

# Call the method using the object
s1.talk()
```

# Inheritance in OOPS :
## Inheritance is an Object-Oriented Programming concept where a new class is created from an existing class. The new class, called a subclass (or derived class), automatically acquires all the attributes and methods of the existing class, which is called the superclass (or base class).

## The primary advantage of inheritance is code reusability. It allows you to create new, more specialized classes without having to rewrite common code.
## Syntax: class Subclass(Superclass):
## An object of the subclass can then use the methods and attributes from both its own class and the superclass.

```python
class A:
    a=1
    b=2
    def method1(cls):
        print(cls.a)
        print(cls.b)
class B(A):
    c=3
    def method2(cls):
        print(cls.c)
print(A.a)  
print(A.b)  
print(B.c)  
print(B.a)  # Inherited attribute
A.method1()  
B.method1() # Inherited methods
B.method2()
```

```python
class A:
    a = 1
    b = 2
    
    @classmethod  # Add this decorator to make it a proper class method
    def method1(cls):
        print(cls.a)
        print(cls.b)

class B(A):
    c = 3
    
    @classmethod  # Add this decorator to make it a proper class method
    def method2(cls):
        print(cls.c)

print(A.a)  
print(A.b)  
print(B.c)  
print(B.a)  # Inherited attribute
A.method1()  # Now works correctly as a class method
B.method1()  # Inherited methods
B.method2()
```

# Polymorphism in OOPS:
## Polymorphism is an Object-Oriented Programming concept that allows a variable, object, or method to take on many different forms or exhibit different behaviors in different contexts.

## The word itself comes from two Greek words: "poly" (meaning many) and "morphos" (meaning forms).
## The central idea of polymorphism is that a single interface can be used for different types of data. For example, the addition operator (+) behaves differently depending on the context:

## When used with numbers, it performs mathematical addition.

## When used with strings, it performs concatenation.

## This ability to perform different actions based on the data type is a form of polymorphism.

```python
def add(a,b):
    return a+b
print(add(10,20))
print(add('manoj','sharma'))
```
