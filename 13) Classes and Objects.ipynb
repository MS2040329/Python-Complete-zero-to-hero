# Types of variables:
## 1) __Instance Variable__: Instance variables are the variables whose separate copy is created in every instance (or object). 
## __Key Characteristic__: A change to an instance variable in one object does not affect the same variable in other objects.
## __How to Define__: They are typically defined inside the constructor method (__init__) using the self keyword (e.g., self.name = "Vishnu").
## 2) __Class/Static Variables__:  class variables are the variables whose single copy is available to all the instances of the class.
## __Key Characteristic__: A change to a class variable is reflected across all instances of that class.
## __How to Define__: They are defined directly inside the class, outside of any methods..:

```python
class Sample:
  # This is a class variable, shared by all objects
  x = 10

  def __init__(self):
    # This is an instance variable, unique to each object
    self.y = 20

s1 = Sample()
s2 = Sample()

print(f"s1.x: {s1.x}, s1.y: {s1.y}")
print(f"s2.x: {s2.x}, s2.y: {s2.y}")
print()

s1.y = 99

print(f"s1.x: {s1.x}, s1.y: {s1.y}")
print(f"s2.x: {s2.x}, s2.y: {s2.y}") # s2.y is not affected
print()

Sample.x = 88

print(f"s1.x: {s1.x}, s1.y: {s1.y}") # s1.x is changed
print(f"s2.x: {s2.x}, s2.y: {s2.y}") # s2.x is also changed
```

# Namespace:
## Namespace is a system that maps names to object. It is a memory block where each name(variable,function,class) you create is linked to specific object.
## This allows you to use same name for different variables in different parts of a program without causing conflicts.
## There are two namespaces:
## 1) __class namespace__: In the class namespace, the names are mapped to class variables.

## 2) __instance variables__:  In the instance namespace, the names are mapped to instance variables.
.#

```python
class Student:
  # This is in the CLASS namespace
  n = 10 

print(f"Original value in class namespace: {Student.n}")

# Create two instances
s1 = Student()
s2 = Student()

# Both instances see the value from the class namespace
print(f"s1 sees: {s1.n}")
print(f"s2 sees: {s2.n}")

# Modify the variable in s1's INSTANCE namespace
s1.n = 99
print(f"s1 now sees: {s1.n}")
print(f"s2 still sees the class value: {s2.n}")
```

# Methods:  
## The purpose of a method is to process the variables provided in the class or in the method.
# Types of methods:
## We can classify variables into three types:
## 1) __instance methods__: Instance methods are the methods that act upon the instance variables of the class. Instance methods are bound to instances (or objects) and are therefore called as: instanceName.method().
## a) __accessor(getter)methods__: Methods that read or "get" the value of instance variables.
## b) __Mutator(setter)methods__: Methods that modify or "set" the value of instance variables.
## 2) __class methods__:  Class methods are the methods which act on the class variables or static variables. These methods are written using __@classmethod__ decorator above them. By default, the first parameter for class methods is cls which refers to the class itself. 
## 3) __static methods__: These are general utility functions that are logically related to a class but do not depend on the state of either the class or a specific instance. These methods are written using __@staticmethod__ decorator above them.
 ods
 

```python
class Student:
    def __init__(self,n=' ',m=0):
        self.name=n
        self.marks=m
    def display(self):
        print('my name is :',self.name)
        print('my marks are :',self.marks)
    # to calculate grades based on marks. 
    def calculate(self): 
        if(self.marks>=600): 
            print('You got first grade') 
        elif(self.marks>=500):
            print('you got second grade')
        elif(self.marks>=350):
            print('you got third grade')
        else:
            print('you failed')
n =int(input('Enter the number of students:'))  

i=0
while i<n:
    name=input('Enter the name of student:')
    marks=int(input('Enter the marks of students'))
    s=Student(name,marks)
    s.display()
    s.calculate()
    i+=1
    
```

```python
class MyClass:
    # Class variable
    class_var = "I am a class variable"

    def __init__(self, name):
        # Instance variable
        self.name = name

    # Instance Method
    def show_name(self):
        print(f"Instance name is: {self.name}")

    # Class Method
    @classmethod
    def show_class_var(cls):
        print(f"Class variable is: {cls.class_var}")

    # Static Method
    @staticmethod
    def utility_function(x):
        print(f"The square of {x} is {x*x}")

# --- Calling the methods ---

# Create an object
obj = MyClass("Manoj")

# Call the instance method on the object
obj.show_name()

# Call the class method on the class
MyClass.show_class_var()

# Call the static method on the class
MyClass.utility_function(5)
```

```python
# class methods
class Bird:
    wings=2
    @classmethod
    def fly(cls,name):
        print('{} flies with {} wings'.format(name,cls.wings))
Bird.fly('sparrow')
Bird.fly('Butterfly')        
```

```python
# static methods:
class Myclass: 
    # this is a class var or a static var 
    n=0    
 
    # constructor that increments n when an instance is created 
    def __init__(self): 
        Myclass.n = Myclass.n+1 
 
    # this is a static method to display the no. of instances 
    @staticmethod 
    def noObjects(): 
        print('No. of instances created: ', Myclass.n) 
 
# create 3 instances 
obj1 = Myclass() 
obj2 = Myclass() 
obj3 = Myclass() 
Myclass.noObjects() 
```

```python
import math
class Sample:
    @staticmethod
    def calculate(x):
        result=math.sqrt(x)
        return result
num=float(input('enter the number'))
res=Sample.calculate(num)
print('The square root of {} is: {:.2f}'.format(num,res))
```

# Passing the members of one class to another class:

## achieved by passing an instance (object) of the first class to a method of the second class.

```python
# This class contains employee details
class Emp:
    def __init__(self, id, name, salary):
        self.id = id
        self.name = name
        self.salary = salary
    
    # Instance method to display details
    def display(self):
        print('Id=', self.id)
        print('Name=', self.name)
        print('Salary= ', self.salary)

class Myclass:
    # Method to receive Emp class instance
    @staticmethod
    def mymethod(e):
        # Access and modify the 'salary' attribute of the Emp object 'e'
        e.salary += 1000
        # Call the 'display' method of the Emp object 'e'
        e.display()

# Main execution
# 1. Create an instance of Emp class
e = Emp(10, 'Manoj', 15000.75)

# 2. Call the static method of Myclass and pass the Emp instance 'e'
Myclass.mymethod(e)
```

# Inner or Nested Classes:

## An inner class is a class defined inside another class. If a class B is written inside class A, then B is considered the inner (or nested) class, and A is the outer class.

## Inner classes are useful when you want to sub-group data or functionality that is strictly related to the outer class. 
## Logical Grouping: For instance, a Person might have a name and age (single values), but a "Date of Birth" involves three values (day, month, year). It makes sense to group these three values into a separate class (Dob) inside the Person class to keep the code organized.

```python
class Person:
    def __init__(self):
        self.name = 'Charles'
        # Create the inner class object here
        self.db = self.Dob() 
    
    def display(self):
        print('Name= ', self.name)
        
    # This is the inner class
    class Dob:
        def __init__(self):
            self.dd = 10
            self.mm = 5
            self.yy = 1988
        def display(self):
            print('Dob= {}/{}/{}'.format(self.dd, self.mm, self.yy))

# Usage
p = Person()
p.display()   # Access outer class method
x = p.db      # Access inner class object created in the constructor
x.display()   # Access inner class method
```
