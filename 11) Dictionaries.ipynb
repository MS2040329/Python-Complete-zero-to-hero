# Dictionaries in Python:
## A dictionary is a Python data structure that stores a group of elements in the form of key-value pairs.  
## Unlike lists or tuples which are indexed by a range of numbers, dictionaries are indexed by keys, which can be any immutable type.
## You can think of a dictionary like a real-world dictionary. You look up a word (the key) to find its definition (the value).

```python
dict = {'Name': 'Chandra', 'Id': 200, 'Salary': 9080.50} 
print(dict)
```

```python
print(dict['Name'])
```

```python
print(dict['Id'])
print(dict['Salary'])
```

# using for loop with dictionary:

```python
colors = {'r': "Red", 'g': "Green", 'b': "Blue", 'w': "White"}
# to retrieve keys 
for k in colors:
    print(k)
```

```python
# to retrieve values
for k in colors:
    print(colors[k])
```

```python
for k,v in  colors.items():
    print('keys:{} and values:{}'.format(k,v))
```

# Sorting the elements of a dictionary in a dictionary

```python
colors = {10: "Red", 35: "Green", 5: "Blue", 15: "White"}
c1= sorted(colors.items(),key=lambda p:p[0])    
print(c1)
c2= sorted(colors.items(),key=lambda p:p[1])
print(c2)
```

# Converting string into dictionary:

```python
lst=[]
my_str = "Vijay=23,Ganesh=20,Lakshmi=19,Nikhil=22" 
for x in my_str.split(','):
    y=x.split('=')
    lst.append(y)
d=dict(lst)
print(d)
```

```python
# to print keys as strings and values as integer
for k,v in d.items():
    d[k]=int(v)
print(d)    
```

# Passing functions to dictionary:

```python
def fun(d):
    for i,j in d.items():
        print(i,'--',j)

fun({'Vijay': 23, 'Ganesh': 20, 'Lakshmi': 19, 'Nikhil': 22})
```

# Ordered dictionary:
## An ordered dictionary is a dictionary but it will keep the order of the elements. The elements are stored and maintained in the same order as they were entered into the ordered dictionary.
## We can create an ordered dictionary using the OrderedDict() method of the ‘collections’ module.le.

```python
from collections import OrderedDict
d=OrderedDict()
d[10] = 'A' 
d[11] = 'B' 
d[12] = 'C' 
d[13] = 'D' 
for i,j in d.items():    
    print(i,j)
```
