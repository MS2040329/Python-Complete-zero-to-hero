# what is file:
## Data is very important. Every organisation depends on its data for continuing its business operations.To store data in a computer, we need files.

# Advantages of storing data in files:
##  When the data is stored in a file, it is stored permanently. This means that even though the computer is switched off, the data is not removed from the memory since the file is stored on hard disk or CD.
##  It is possible to update the file data. For example, we can add new data to the existing file, delete unnecessary data from the file and modify the available data of the file. This makes the file more useful. 
##  Once the data is stored in a file, the same data can be shared by various programs. 
##  Files are highly useful to store huge amount of data.

# Types of files in Python:
## __Text Files__: Text files store the data in the form of characters.
## __Binary Files__: Binary files store entire data in the form of bytes, i.e. a group of 8 bits each.
## Binary files can be used to store text, images, audio and video.  Image files are generally available in .jpg, .gif or .png formats. 
##  images contain pixels which are minute dots. Each pixel can be represented by a bit, i.e. either 1 or 0.

# __tell() Method__:
## Tells the current position of the cursor

```python
with open("myfile.txt", "r") as f:
   f.seek(10) # Move cursor to byte position 10
   print(f.tell()) # Output: 10
   print(f.readline()) # Reads from position 10 onwards
```

# __Pickle or Serialisation__: if we want to store some structured data (has different types) in the files.
## Pickle is a process of converting a class object into a byte stream so that it can be stored into a file. This is also called object serialization. Pickling is done using the dump() method of ‘pickle’ module .

# __Unpickle or Deserialisation__:
## Unpickle is a process whereby a byte stream is converted back into a class object. It means, unpickling represents reading the class objects from the file. Unpickling is also called desearialization.  Unpickling is done using the load() method of ‘pickle’ module


```python
import pickle
data = {'a': [1, 2.0, 3, 4+6j], 'b': ("character string", b"byte string"), 'c': {None, True, False}}
# Pickling the dictionary
with open('data.pkl','wb')as f:
    pickle.dump(data,f)
# Unpickling the dictionary
with open('data.pkl','rb')as f:
    data_loaded=pickle.load(f)
print(data_loaded)
```

```python
class Emp:
    def __init__(self,name,age,sal):
        self.name=name
        self.age=age
        self.sal=sal
    def display(self):
        return f"{self.name} is {self.age} years old and earning {self.sal} rupees"
emp=Emp("Manoj",20,100000)
```

```python
with open('data.pkl','ab')as f:
    pickle.dump(emp,f)
# Unpickling the dictionary
with open('data.pkl','rb')as f:
    data_loaded=pickle.load(f)
print(emp.display())
```

```python
# creating a binary file and storing a group of strings
# view binary files as a record of fixed length
record_length=20
with open("cities.bin", "wb") as f: 
    n=int(input("How many entries: "))
    for i in range(n):
        city=input("Enter the name of the cities")
        ln=len(city)
        city=city+(20-ln)*' '
        city=city.encode() # convert into byte format
        f.write(city)
        
```

```python
import os
#length of each record in file
record_length=20

# finding the size of the file
size=os.path.getsize('cities.bin')
print(f"The size of the file is {size}")

# finding the no of records in the file
records=int(size/record_length)
print(f"There are {records} records in file")

with open("cities.bin","rb") as f:
    name= input("Enter the name of the cities")
    name= name.encode()
    position= 0    
    found= False

    for i in range(records):
        f.seek(position)
        str=f.read(20)
    
        if name in str:
            print(f"match found at record no {i+1}")
            found=True
        position+=record_length

if not found:
    print("City not found")
```

```python
# Update or modify the record in the binary file
import os
record_length=20
# size of the file
size=os.path.getsize("cities.bin")
print(f"The size of binary file is {size}")
# records in the file
records=int(size/record_length)
print(f"The no of records in file {records}")

with open("cities.bin","r+b") as f:
    name=input("Enter the name of the cities: ")
    name=name.encode()

    new_name=input("Enter the new name")
    ln=len(new_name)
    new_name=new_name+(20-ln)*" "
    new_name=new_name.encode()

    position=0
    found=False

    for i in range(records):
        f.seek(position)
        str=f.read(20)
        if name in str:
            print(f"updated records:{i+1}")
            found=True
            f.seek(-20,1)
            f.write(new_name)
        position+=record_length
    if not found:
        print("city not found")
```

```python
# Delete a specific record from the binary file
import os
record_length=20
size=os.path.getsize("cities.bin")
records=int(size/record_length)
f1=open("cities.bin", 'rb')
f2=open("file2.bin",'wb')
city=input("Enter the city name to delete")
ln=len(city)
city=city+(20-ln)*' '
city=city.encode()
for i in range(records):
    str=f1.read(record_length)
    if(str!=city):
        f2.write(str)
print("record deleted!")
f1.close()
f2.close()
# delete the cities.bin file 
os.remove("cities.bin") 
# rename file2.bin as cities.bin 
os.rename("file2.bin", "cities.bin")
```

# Random Access of binary file using mmap:

## __mmap__ – ‘memory mapped file’ is a module in Python that is useful to map or link to a binary file and manipulate the data of the file as we do with the strings. It means, once a binary file is created with some data, that data is viewed as strings and can be manipulated using mmap module. The first step to use the mmap module is to map the file using the mmap() method as:

## Best for large binary files or inter-process communication.
## Array-like (slicing/indexing).
## Faster (OS handles paging).

```python
import mmap
with open("project_data.bin", "wb") as f:
    f.write(b"I am learning core python concepts")
with open("project_data.bin", "r+b") as f:
    with mmap.mmap(f.fileno(),0) as mm: # indicates that fileno() is a handle to the file object ‘f’.  
#zero( 0 ) represents the total size of the file should be considered for mapping.        
        print(mm.read())
        print(mm.readline())
        print(mm[5:])   # display from 5th byte till the end 
        print(mm[5:10])   # display from 5th to 9th bytes 
        mm[0:5] = b"manoj"  #Modify the binary file
        n=mm.find(b"learning")
        f.write(b"python"+b"learning")
```

```python
# Python program to create a phonebook with names and numbers
with open("phonebook.dat","wb") as f:
    n=int(input("How many entries in the phone book"))
    for i in range(n):
        name=input("Enter the name")
        phone=input("Enter the phone number")
        name=name.encode()
        phone=phone.encode()
        f.write(name+phone)
```
