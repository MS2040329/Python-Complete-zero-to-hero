# Comments in Python:
Comments are non-executable statements in a Python program that are used to describe the code and improve its readability. The Python compiler and the Python Virtual Machine (PVM) ignore comments, so they do not affect the program's execution.
## Types of Comments:
Python supports two main types of comments:

### Single-line comments:
These start with the hash symbol (#). Any text following the # on that line is treated as a comment. They are useful for adding brief explanations or disabling a single line of code.

# This is a single-line comment
a = 10     # This comment explains the purpose of the line

### Multi-line comments (Docstrings):
While Python does not have a dedicated syntax for multi-line comments, programmers use triple quotes (""" or ''') to create them. When a string enclosed in triple quotes is placed as the first statement in a module, class, or function, it is called a "docstring". 
#### Docstrings:
Docstrings are used to provide detailed documentation and can be accessed by tools to automatically generate API documentation. If these strings are not assigned to a variable, they are ignored by the interpreter and can serve as de facto multi-line comments.

# variables in Python:
In Python, a variable is a name or tag that points to an object in memory, rather than a fixed memory location that stores a value directly. This approach is different from other languages like C or Java, where variables are typically seen as containers or storage boxes

Variables are assigned in Python using the = operator.

When you assign a value to a variable, such as a = 1, 
Python first creates an object for the value 1 and then attaches the name a to that object.

If you later reassign the variable a = 2, Python creates a new object for the value 2 and moves the a tag to point to this new object. The old object (in this case, 1) remains in memory until it's no longer referenced by any name, at which point the garbage collector can remove it.

This design makes Python memory-efficient, as multiple variable names can point to the same object.

For instance, if you set b = a, a new tag b is created to point to the exact same object that a points to, rather than creating a separate copy of the data.This is called shared reference.

# Datatypes in Python
In Python, a datatype is a classification that specifies which type of value a variable can hold. Python's type system is dynamic, so you don't need to declare a variable's datatype explicitly; the interpreter infers it automatically at runtime based on the value assigned.





Python has two categories of datatypes:

## Built-in Datatypes
These are the data types that come standard with the Python language. They are:


### None Type: 
A special type that represents an object with no value, similar to null in other languages.In a Python program, maximum of only one ‘None’ object is provided.
 In Boolean 
expressions, ‘None’ datatype represents ‘False’.   

### Numeric Types: 
These represent numbers and include int (for integers), float (for decimal numbers), and complex (for numbers with real and imaginary parts).

### Sequences:
These are ordered collections of elements. They include str (for strings), bytes, bytearray, list, tuple, and range.

### Bool datatype:
The bool data type in Python represents boolean values, which can only be either True or False. Internally, Python represents True as 1 and False as 0. This data type is fundamental for control flow, as conditions in if, while, and other statements are evaluated to a Boolean value 

### Sets: 
Unordered collections of unique elements, including set and frozenset.

### Mappings:
These store data in key-value pairs, with dict (for dictionaries) being the primary example.

## User-defined Datatypes:
These are complex data types created by programmers, such as classes, arrays, or modules.

# Identifiers:
In Python, identifiers are names used to identify variables, functions, classes, modules, or other objects. They are the labels you create to reference different elements in your code. 

The rules for naming an identifier are:
1) It can include letters (A-Z, a-z), numbers (0-9), and the underscore character (_).
2) It must start with a letter or an underscore, not a number.

3) It is case-sensitive, meaning salary, Salary, and SALARY are all considered different identifiers.
4) Special symbols like ?, #, $, and @ are not allowed.
# Reserved keywords:
Reserved words, on the other hand, are a specific set of words that have a predefined meaning in Python and cannot be used as identifiers.


# Naming convention in Python:
In Python, naming conventions are a set of rules and suggestions for how to name various programming components to ensure readability and consistency. Adhering to these conventions, often referred to as PEP 8, makes code easier for you and others to understand and maintain.

# Key Naming Conventions
## Packages and Modules: 
Use all lowercase letters with words separated by underscores for multi-word names (e.g., my_package, my_module.py).

## Classes:
Class names should use CamelCase, where each word starts with a capital letter (e.g., MyClass). For exception classes, the name should end with the word Error (e.g., MyCustomError).

## Functions, Methods, and Variables:
Use all lowercase letters for names, and separate words with underscores (e.g., my_function, instance_variable).

## Constants:
Constant names should be in all capital letters, with underscores separating words (e.g., MAX_SIZE).

## Method Arguments:
The first argument of an instance method should always be self, and the first argument of a class method should be cls.

## Non-public members:
Names that are not meant to be accessed from outside a class should start with a single underscore (e.g., _non_public_method). Members that are not accessible at all should have two underscores before and after their names (e.g., __init__).
