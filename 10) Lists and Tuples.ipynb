# Indexing in list:
## Indexing retrieves a single element from a list using its position number. The index is specified in square brackets [], and the first element is always at index 0.

```python
my_list=[10,20,30,40,'manoj',49.5,12.6]
print(my_list[4])
print(my_list[0])
print(my_list[-1])
print(my_list[-2])
```

# Slicing in list:
## Slicing creates a new list by extracting a range of elements from an existing list. The syntax is list[start:stop:step].

```python
my_list = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
print(f"Original list: {my_list}\n")
sub_list1 = my_list[1:4]
print(f"Slice [1:4]: {sub_list1}")

sub_list2 = my_list[:3]
print(f"Slice [:3]:  {sub_list2}")

sub_list3 = my_list[3:]
print(f"Slice [3:]:   {sub_list3}")

sub_list4 = my_list[::2]
print(f"Slice [::2]: {sub_list4}")

copied_list = my_list[:]
print(f"Slice [:]:   {copied_list}")
```

# Modifying lists: 
## lists are mutable, you can change their contents after they have been created. This includes updating existing elements, adding new ones, and deleting elements.

```python
my_list = [10, 20, 30, 40, 50]
print(f"Original list:           {my_list}")

my_list[1] = 99
print(f"After updating index 1:  {my_list}")

my_list.append(60)
print(f"After appending 60:      {my_list}")

del my_list[2]
print(f"After deleting index 2:  {my_list}")
```

# Finding the minimum and maximum elements in a list:
## max() and min(), which return the biggest and smallest elements from a list..

```python
x=[10,20,30,40,50,60]
y=min(x)
z=max(x)
print(y)
print(z)
```

```python
x = [20, 10, 5, 20, 15]

# Assume the first element is both the biggest and smallest
big = x[0]
small = x[0]

# Loop through the rest of the list
for i in range(1, len(x)):
    if x[i] > big:
        big = x[i]
    if x[i] < small:
        small = x[i]

print(f"The list is: {x}")
print(f"Biggest number is: {big}")
print(f"Smallest number is: {small}")
```

# Sorting a list using bubble sort:

```python
x=[]
n=int(input('how many elements'))
for i in range(n):
    x.append(int(input()))
print(x)
for i in range(n-1): # for no of passes
    for j in range(n-1-i): # for comparison in each pass
        if x[j]>x[j+1]:
            t=x[j]
            x[j]=x[j+1]
            x[j+1]=t
print(f'sorted list {x}')       
```

# List comprehension in Python:
## List comprehensions represent the creation of new lists from an iterable object (like a list, set, tuple, dictionary, or range) that satisfy a given condition. List comprehensions contain very compact code, usually a single statement that performs the task. 
### basic syntax: [expression for item in iterable if condition]. 

```python
# traditional for loop method:
square=[]
for i in range(0,11):
    square.append(i**2)
print(square)
```

```python
# using list comprehension mathod:
even_squares=[x**2 for x in range(0,11) if x%2==0]
print(even_squares)
```

```python
x=[10,20,30]
y=[1,2,3,4]
lst=[]
for i in x:
    for j in y:
        lst.append(i+j)
print(lst)
```

```python
# doing the same above thing using a list comprehension
lst=[i+j for i in [1,2,3] for j in [4,5,6]]
print(lst)
```

```python
lst = [ i+j for i in 'ABC' for j in 'DE'] 
print(lst)
```

```python
words = ['Apple', 'Grapes', 'Banana', 'Orange'] 
lst=[w[0] for w in words]
print(lst)
```

```python
num1=[1,2,3,4,5,6]
num2=[10,20,30,1,2]
lst=[i for i in num1 if i not in num2]
print(lst)
```

```python
# dictionary comprehension
dict = {1001: 'Pratap', 1002: 'Mohan', 1003: 'Ankita'} 
dict1={value:key for key,value in dict.items()}
print(dict1)
```
