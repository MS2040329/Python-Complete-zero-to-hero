# Comparing strings:
## You can compare strings in Python using the standard relational operators: ==, !=, <, >, <=, and >=. These operators return a Boolean value


```python
s1='Box' 
s2='Boy' 
if(s1==s2): 
    print('Both are same') 
else: 
    print('Not same') 
```

```python
s1 = 'Box'
s2 = 'Boy'
s3 = 'Box'
s4 = 'box'

# Equality check
print(f"'{s1}' == '{s3}': {s1 == s3}")  

# Inequality check
print(f"'{s1}' != '{s2}': {s1 != s2}")  

# Alphabetical order check
# 'Box' comes before 'Boy' in the dictionary
print(f"'{s1}' < '{s2}': {s1 < s2}")    

# Case-sensitivity check
# 'Box' (uppercase B) is considered less than 'box' (lowercase b)
print(f"'{s1}' == '{s4}': {s1 == s4}")  
print(f"'{s1}' < '{s4}': {s1 < s4}")    
```

## Strings are immutable(why?)
1. __performance__:  When an object is immutable, it will have a fixed size in memory since it cannot be modified. Because strings are immutable, we can easily allocate memory space for them at creation time, and the storage requirements are fixed and unchanging. Hence, it takes less time to allocate memory for strings and less time to access them. This increases the performance of the software.

string object will create a new object in memory. Thu,s the identity number of the new
object will chan,gwhichat lets the programmeo understand that somebody modified 
the original string. This is useful to enforce security where strings can be transportd

# searching in strings:

```python
# A list of strings to search through
str_list = ['Ramesh', 'Kumar', 'Sriya', 'Vinod', 'Ramesh']

# The string to search for
search_term = 'Ramesh'

# A flag to track if the term was found
found = False

# Loop through the list with index
for i in range(len(str_list)):
    # Check if the current string matches the search term
    if search_term == str_list[i]:
        print(f"Found '{search_term}' at position: {i+1}")
        found = True

# If the loop finishes and the flag is still False, the term was not found
if not found:
    print(f"'{search_term}' was not found in the list.")
```

```python
original_string = 'It is summer'
substring_to_insert = 'hot'
position = 7 # The index where the new substring will start
start = original_string[:position]
end = original_string[position:]
# Concatenate the three parts to create the new string
new_string = start + substring_to_insert + end
print(f"Original string: '{original_string}'")
print(f"New string: '{new_string}'")
```
