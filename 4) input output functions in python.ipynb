# Output statements: Print() function
To display output or results, Python provides the print() function. This function can be 
used in different formats

```python
#syntax: print("string") or print('string')
print("manoj sharma")
#using print() alone through the cursor to the next line
print() 
print("btech ai and ds")
```

```python
# using escape characters in print() statements
print("manoj\nsharma")
print("manoj\tsharma")
print("manoj\\nsharma")
# Repetition
print("manoj\t"*5)
# concatenation
print("manoj"+"sharma")
print("manoj","sharma")
```

```python
# syntax:print(variable list)
a,b=20,30
print(a,b)
print(a,b,sep='----')
print(a,b,sep='*****')
```

#### We can ask the print() function not to throw the cursor into the next line but display the output in the same line. This is done using end attribute  


```python
print('manoj',end=' ')
print('sharma')
print('manoj',end='\t')
print('sharma')
```

```python
# syntax: print(object)
lst=[4.5,5,9,'manoj']
print(lst)
d = {'Idly':30.00, 'Roti':45.00, 'Chappati':55.50} 
print(d) 
```

```python
# syntax : print("string",variables)
a=2
b=20
print(a,b, "is even number")  
print('You typed ', a,b, 'as input')  
```

# Specifier	      Description                
%s	         Formats as a string.	
%d or %i	 Formats as a decimal integer.	
%f or %F	 Formats as a floating-point number. You can specify precision, like %.2f. 
%c	        Formats as a single character.	
%x or %X	Formats as a hexadecimal number.	
%o	        Formats as an octal number.	
%e or %E	Formats a floating-point number in scientific notation.	
%%	         Displays a literal percent sign.

```python
# syntax: print(formatted string)
x, y,z,m = 10, 15.68,16,'manoj' 
print('x= %i y= %f z=%d m=%s' % (x, y,z,m)) 
```

```python
name='manoj' 
print('Hai %s' % name) 
# aligns string to the right
print('Hai (%20s)' % name) 
# aligns string to the left
print('Hai (%-20s)' % name)
# %c used to display a single character.
print('Hai %c, %c' % (name[0], name[1]))
print('Hai %s' %(name[0:2]))
```

# .format() method:
We can use a replacement field, which is denoted by a pair of 
curly braces { }. We can mention names or indexes in these replacement fields. These 
names or indexes represent the order of the value
<hr>s.
syntax: print(‘format string with replacement fields’.format(values))


```python
m,n="manoj",'ai and ds'
print("my name is {0}. i am persuing btech {1}".format(m,n))
```

```python
print('manoj is a{}'.format(' good boy'))
```

```python
print('the{} {} {}'.format(' nature','is','the best healer'))
```

```python
print('the {2} {0} {1} ' .format('brown','fox','quick'))
```

```python
print('the {q} {b} {f} '.format(f='fox',b='brown',q='quick'))
print('the{f} {f} {f} '.format(f=' fox'))
```

```python
print('the result is {r:10.6f}'.format(r=100/99))
```

```python
print("Name: {:s}, Salary: {:.2f}".format("Bob", 12345.678))
```
